```python
# Import the necessary libraries
import nltk
from bs4 import BeautifulSoup as bs
import numpy as np
import pandas as pd
import requests as rq

# Import libraries for creating a word cloud
import matplotlib.pyplot as plt
from nltk.corpus import stopwords

# John Wick: Chapter 4 reviews
# Assigning the IMBD URL and movie ID
imbd_url = 'https://www.imdb.com/title/'  # IMDb base URL
movie_id = 'tt10366206'  # Specific movie ID

# Create a DataFrame with one column to store the text of reviews
df = pd.DataFrame(columns=['Review'])

# Number of pages to web scrape
number_pages_webscrap = 50  

# Loop through the pages to scrape user reviews
for page in range(1, number_pages_webscrap + 1):
    # Constructing the URL for the current page
    url = f'{imbd_url}{movie_id}/reviews?start={((page - 1) * 10)}'
    
    # Sending a GET request to the IMDb page and retrieve the content as text
    req = rq.get(url).text
    
    # Parsing the HTML content using BeautifulSoup
    s = bs(req, 'html.parser')
    
    # Finding all user reviews on the page
    user_review = s.find_all('div', attrs={'class': 'text show-more__control'})
    
    # Creating an empty list to store the reviews
    reviews_list_collection = []
    
    # Extracting and append each user review to the list
    for i in range(len(user_review)):
        reviews_list_collection.append(user_review[i].get_text())
    
    # Concatenating the reviews to the DataFrame
    df = pd.concat([df, pd.DataFrame({'Reviews': reviews_list_collection})], ignore_index=True)

# Calculating the word count for each review and add it as a new column
df['User_word_count'] = df['Reviews'].apply(lambda x: len(str(x).split(" "))

# printing head of df
print(df[['Reviews', 'User_word_count']].head())
```

```python
# Write your code here
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import string

# Load the CSV file
# Read the CSV file containing the movie reviews into a DataFrame
df1 = pd.read_csv("John Wick_Chapter 4_reviews.csv")

# Function to preprocess and clean review
def review_preprocessing(review):
    # Remove punctuation and special characters
    review = ''.join([character for character in review if character not in string.punctuation])
    
    # Remove numbers
    review = ''.join([character for character in review if not character.isdigit()])
    
    # Tokenize the review
    words = nltk.word_tokenize(review)
    
    # Remove stopwords
    words = [w for w in words if w.lower() not in stopwords.words('english')]
    
    # Lowercase all words
    words = [w.lower() for w in words]
    
    # Stemming
    stemmer = PorterStemmer()
    words = [stemmer.stem(w) for w in words]
    
    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(w) for w in words]
    
    # Join the words back into a cleaned sentence
    cleaned_user_reviews = ' '.join(words)
    
    return cleaned_user_reviews

# Apply the preprocessing function to the "Reviews" column and create a new column
# Create a new column in the DataFrame containing the cleaned and preprocessed reviews
df1['Reviews_After_Cleaning'] = df1['Reviews'].apply(review_preprocessing)
```

### Constituency Parsing Tree:

A sentence's grammatical structure may be represented in a hierarchical fashion by using something called a constituency parsing tree. It does this by dismantling a sentence into its component phrases, which include noun phrases (NP), verb phrases (VP), prepositional phrases (PP), and clauses. Each node in the tree represents a component phrase, and the edges show how these phrases are nested inside each other at various levels. As you progress down the tree, you will eventually reach individual words before you reach the root node, which often symbolizes the full sentence. As you move away from the root node, you will face smaller and smaller phrases. Each phrase has a name indicating the grammatical type it falls under (such as NP or VP), and it may include words or other phrases inside it.



### Dependency Parsing Tree:

A dependency parsing tree is used to describe the syntactic connections that exist between the words in a phrase. Each word is represented as its own node in this form of tree, and the connections between the nodes show how the words are connected to one another in terms of their grammatical and syntactic dependencies. In a dependency tree, there are normally two different kinds of nodes: the root node, which represents the most important verb or main phrase of the sentence, and dependent nodes, which represent additional words that rely on the root or on each other. The root node is always at the beginning of the tree.
